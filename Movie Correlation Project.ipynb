```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read in the data

df = pd.read_csv(r'C:\Users\madei\Downloads\sql things\Case Studies\Movie Industry\archive\movies.csv')


```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('float64', errors='ignore')
df['gross'] = df['gross'].astype('float64', errors='ignore')
```

```python
df.head()
```

```python
# Create correct Year colum

df['yearcorrect'] = df['year'].astype(str).str[:4]
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')


plt.show()





```

```python
df.head()
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Let's start looking at correlation
# pearson, kendall, spearmanplt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

df.corr(method='pearson')
```

```python
# High correclation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
# Looks at Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
df_numerized.corr().head()
```

```python
correlation_nat = df_numerized.corr()

corr_pairs = correlation_nat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```
